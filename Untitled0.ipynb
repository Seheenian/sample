```python
!pip install spacy
!python -m spacy download en_core_web_lg
import spacy
import en_core_web_lg
nlp = en_core_web_lg.load()
```

```python
# Lemmatizing
def lemmatize(reddit): 
    words = ' '.join(reddit)
    doc = nlp(words)
    return [token.lemma_ for token in doc]
reddit = lemmatize(reddit)
```

```python
d = []
nlp = en_core_web_lg.load()
doc = nlp(" ".join(reddit))
for token in doc:
    d.append((token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
          token.shape_, token.is_alpha, token.is_stop))

reddit_pd = pd.DataFrame(d, columns=("text", "lemma", "pos", "tag", "dep", "shape", 
                               "is_alpha", "is_stop"))
reddit_pd
```

```python
# Removing Stopwords
stop2 = list(np.where(reddit_pd["is_stop"] == False)[0])
reddit = reddit_pd.iloc[stop2]
```

```python
# Counting the Frequencies
reddit_freq = reddit["text"].value_counts()
```

```python
# Transforming to Dictionary and Removing Additional Punctuations and Type errors
reddit_freq = {k:v for k, v in reddit_freq.iteritems()}
try: 
    del reddit_freq["-PRON-"]
    del reddit_freq["·"]
    del reddit_freq["iso"]
    del reddit_freq["al"]
    del reddit_freq["bs"]
    del reddit_freq["din"]
    del reddit_freq["et"]
    del reddit_freq["s"]
    del reddit_freq['"']
    del reddit_freq["’"]
    del reddit_freq["'"]
    del reddit_freq["◦"]
    del reddit_freq["−"]
    del reddit_freq["."]
except KeyError: 
    print("key item not found")
```

```python
# Wordcloud Visualization
wordcloud = WordCloud(max_font_size=60, max_words=100, background_color="white", colormap = "PuBuGn_r", random_state=None, ranks_only=False).generate_from_frequencies(reddit_freq)
plt.figure(figsize = (18, 18))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file("drive/My Drive/DIGHUM/plastic2010_wordcloud.png")
```
